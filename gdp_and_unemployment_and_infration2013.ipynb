```python
import pandas as pd
```

```python
data = pd.read_csv("data_Data.csv", header=None)
```

```python
data2 = data [[2,50]]
#国名と、2013年度分を抜き取ります
data3 = data2.ix[21:40]
#GDP成長率の列だけ抜き出します21〜40まで
data4 = data3.rename(columns={2:'Country',50: 'GDP growth2013(%)'})
#カラムを番号から国名とGDP成長率に変更します
data4
```

```python
rename_row={x+21:x for x in range(29)}
#インデックスを０からの数字に変更するための辞書をつくります #これを変数としてしまっておきます
data5 = data4.rename(index=rename_row)
#インデックス変換です
a = [i for i in data5["Country"]]
#インデックスの数字をやっぱり国名にしておきたいと思ったら、国名をリストにしておきます
newindexes = {x:y for y,x in zip(a, range(29))}
#これで二重ループのリスト内包表記をすればrename用の辞書ができあがります
#変数に格納しておきます
data6 = data5[["GDP growth2013(%)"]].astype(float)
#GDP growthを浮動小数点に変更しましょう
data7 = data6.rename(index=newindexes)
#さらにここでインデックスを国名に変えます
data7
```

```python
datav2_2 = data [[2,50]]
#国名と、2013年度分を抜き取ります
datav2_3 = datav2_2.ix[121:140]
#失業率（Unemployment, total (% of total labor force)）の列だけ抜き出します121〜140まで
datav2_4 = datav2_3.rename(columns={2:'Country',50: 'Unemployment,2013(%)'})
#カラムを番号から国名と失業率に変更します
```

```python
rename_row={x+121:x for x in range(20)}
#インデックスを０からの数字に変更するための辞書をつくります #これを変数としてしまっておきます
datav2_5 = datav2_4.rename(index=rename_row)
#インデックス変換です
a = [i for i in datav2_5["Country"]]
#インデックスの数字をやっぱり国名にしておきたいと思ったら、国名をリストにしておきます
newindexes = {x:y for y,x in zip(a, range(20))}
#これで二重ループのリスト内包表記をすればrename用の辞書ができあがります
#変数に格納しておきます
datav2_6 = datav2_5[["Unemployment,2013(%)"]].astype(float)
#Unemployment,2014(%)を浮動小数点に変更しましょう
datav2_7 = datav2_6.rename(index=newindexes)
#さらにここでインデックスを国名に変えます
datav2_7
```

```python
datav3_2 = data [[2,50]]
#国名と、2013年度分を抜き取ります
datav3_3 = datav3_2.ix[81:100]
#インフレ率（Inflation, consumer prices (annual %)）の列だけ抜き出します８1〜1０0まで
datav3_4 = datav3_3.rename(columns={2:'Country',50: 'Inflation, consumer prices,2013(%)'})
#カラムを番号から国名と失業率に変更します

rename_row={x+81:x for x in range(20)}
#インデックスを０からの数字に変更するための辞書をつくります #これを変数としてしまっておきます
datav3_5 = datav3_4.rename(index=rename_row)
#インデックス変換です
a = [i for i in datav3_5["Country"]]
#インデックスの数字をやっぱり国名にしておきたいと思ったら、国名をリストにしておきます
newindexes = {x:y for y,x in zip(a, range(20))}
#これで二重ループのリスト内包表記をすればrename用の辞書ができあがります
#変数に格納しておきます
datav3_6 = datav3_5[["Inflation, consumer prices,2013(%)"]].astype(float)
#Unemployment,2014(%)を浮動小数点に変更しましょう
datav3_7 = datav3_6.rename(index=newindexes)
#さらにここでインデックスを国名に変えます
datav3_7
```

```python
concat1 = pd.concat([data7, datav2_7], axis=1)
#axis=1横方向に連結！
concat2 = pd.concat([concat1, datav3_7], axis=1)
#axis=1横方向に連結！
concat2
```

```python
#ここで描写ツールをインポートしておきましょう
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline　
#matplotlibがインライン（このブラウザ内で表示されるという意味です。）になるマジックコマンド
```
