```python
#pandasで散布図をつくるまで

import pandas as pd
#まずはpandasを読み込みましょう
```

```python
data = pd.read_csv("data_Data.csv", header=None)
#csvDataを読み込みましょう
```

```python
data
```

```python
data2 = data [[2,51]]
#国名と、2014年度分を抜き取ります
data3 = data2.ix[21:40]
#GDP成長率の列だけ抜き出します21〜40まで
data4 = data3.rename(columns={2:'Country',51: 'GDP growth2014(%)'})
#カラムを番号から国名とGDP成長率に変更します
```

```python
rename_row={x+21:x for x in range(29)}
#インデックスを０からの数字に変更するための辞書をつくります #これを変数としてしまっておきます
data5 = data4.rename(index=rename_row)
#インデックス変換です
a = [i for i in data5["Country"]]
#インデックスの数字をやっぱり国名にしておきたいと思ったら、国名をリストにしておきます
newindexes = {x:y for y,x in zip(a, range(29))}
#これで二重ループのリスト内包表記をすればrename用の辞書ができあがります
#変数に格納しておきます
data6 = data5[["GDP growth2014(%)"]].astype(float)
#GDP growthを浮動小数点に変更しましょう
data7 = data6.rename(index=newindexes)
#さらにここでインデックスを国名に変えます
```

```python
data7
```

```python
datav2_2 = data [[2,51]]
#国名と、2014年度分を抜き取ります
datav2_3 = datav2_2.ix[121:140]
#失業率（Unemployment, total (% of total labor force)）の列だけ抜き出します121〜140まで
datav2_4 = datav2_3.rename(columns={2:'Country',51: 'Unemployment,2014(%)'})
#カラムを番号から国名と失業率に変更します
```

```python
datav2_4
```

```python
rename_row={x+121:x for x in range(20)}
#インデックスを０からの数字に変更するための辞書をつくります #これを変数としてしまっておきます
datav2_5 = datav2_4.rename(index=rename_row)
#インデックス変換です
a = [i for i in datav2_5["Country"]]
#インデックスの数字をやっぱり国名にしておきたいと思ったら、国名をリストにしておきます
newindexes = {x:y for y,x in zip(a, range(20))}
#これで二重ループのリスト内包表記をすればrename用の辞書ができあがります
#変数に格納しておきます
datav2_6 = datav2_5[["Unemployment,2014(%)"]].astype(float)
#Unemployment,2014(%)を浮動小数点に変更しましょう
datav2_7 = datav2_6.rename(index=newindexes)
#さらにここでインデックスを国名に変えます
datav2_7
```

```python
concat1 = pd.concat([data7, datav2_7], axis=1)
#axis=1横方向に連結！
```

```python
concat1
```

```python
#ここで描写ツールをインポートしておきましょう
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline　
#matplotlibがインライン（このブラウザ内で表示されるという意味です。）になるマジックコマンド
```

```python
sns.lmplot("GDP growth2014(%)","Unemployment,2014(%)",concat1,fit_reg=False)
#seabornで散布図ができましたね^^
```

```python
sns.lmplot("GDP growth2014(%)","Unemployment,2014(%)",concat1)
#散布図に回帰直線を書いてみました
```

```python
sns.jointplot("GDP growth2014(%)","Unemployment,2014(%)",concat1,kind='scatter',color='indianred')
#jointplotを描いてみました
```

```python
concat1.corr()
#相関係数の計算をしてくれるメソッドです
#相関係数は絶対値0.2を超えてきているのでややある程度ですね。
```

```python
sns.heatmap(concat1.corr(), annot=True)
#ヒートマップにしてみましょう
```
