```python
#pandasで分布の横棒グラフをつくるまで

import pandas as pd
#まずはpandasを読み込みましょう
```

```python
data = pd.read_csv("data_Data.csv", header=None)
#csvDataを読み込みましょう
```

```python
data[:5]
#最初から数えて5個分を出してみます
```

```python
%matplotlib inline
#matplotlibがインライン（このブラウザ内で表示されるという意味です。）になるマジックコマンド
```

```python
data2 = data [[2,52]]
#国名と、2015年度分を抜き取ります
```

```python
data2
```

```python
data3 = data2.ix[21:40]
#GDP成長率の列だけ抜き出します21〜40まで
```

```python
data3
```

```python
data4 = data3.rename(columns={2:'Country',52: 'GDP growth'})
#カラムを番号から国名とGDP成長率に変更します
```

```python
data4
```

```python
{x+21:x for x in range(29)}
#インデックスを０からの数字に変更するための辞書をつくります
```

```python
rename_row={x+21:x for x in range(29)}
#これを変数としてしまっておきます
```

```python
data5 = data4.rename(index=rename_row)
#インデックス変換です
```

```python
data5
```

```python
a = [i for i in data5["Country"]]
#インデックスの数字をやっぱり国名にしておきたいと思ったら、国名をリストにしておきます
```

```python
a
```

```python
{x:y for y,x in zip(a, range(29))}
#これで二重ループのリスト内包表記をすればrename用の辞書ができあがります
```

```python
newindexes = {x:y for y,x in zip(a, range(29))}
#変数に格納して
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
#ちょっとここで描写系のパッケージを入れ込んでおきましょう
```

```python
data5.dtypes
#で、いまplot()メソッドをしてもエラーがでます。その理由GDPgrowthの型がobjectで、数値型ではないから
```

```python
data6 = data5[["GDP growth"]].astype(float)
#なので、浮動小数点に変更しましょう
```

```python
data6
```

```python
data7 = data6.rename(index=newindexes)
#さらにここでインデックスを国名に変えます
```

```python
data7
```

```python
data7.plot(kind='barh', color='r', alpha=0.6)
#さあ、プロットしましょう　barh=横棒のグラフ、color=棒の色、alphaは棒の色の透明度です
```

```python
data7.plot(kind='bar', color='b', alpha=0.6)
#縦型(bar)だとこんな感じですね
```
